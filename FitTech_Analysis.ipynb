**2. ზოგადი შედეგები**

ქვემოთ მოცემული კოდის საშუალებით გავიგებთ, აქტიური მომხმარებლების საერთო რაოდენობასა და ჩატარებული ვარჯიშების სესიების რაოდენობას.

```python
import pandas as pd

df = pd.read_csv('/content/sample_data/fittech_data.csv')

total_active_users = df['მომხმარებლის ID'].nunique()
total_workouts = len(df)

print(f"აქტიური მომხმარებლების საერთო რაოდენობა: {total_active_users:,}")
print(f"ვარჯიშის სესიების საერთო რაოდენობა: {total_workouts:,}")
```

შედეგად მივიღეთ, რომ აქტიური მომხმარებლების რაოდენობა უდრი 10000 -ს და სესიების საერთო რაოდენობაა 477481.

**3. ვარჯიშის ტიპების ეფექტურობა**

განვიხილოთ, თუ როგორ შეუწყო ხელი თითოეულმა ვარჯიშის ტიპმა მომხმარებელთა ჩართულობას:

```python
workout_performance =  df.groupby('ვარჯიშის ტიპი').agg(
    საშუალო_ხანგრძლივობა = ('ხანგრძლივობა', 'mean'),
    ჯამური_ხანგრძლივობა = ('ხანგრძლივობა', 'sum'),
    ჯამური_კალორიები = ('დამწვარი კალორიები', 'sum'),
    სესიები = ('ვარჯიშის ტიპი', 'count')
)

workout_performance.columns = ['საშუალო ხანგრძლივობა', 'ჯამური ხანგრძლივობა', 'ჯამური კალორიები', 'სესიები']

workout_performance = workout_performance.sort_values(by='სესიები', ascending=False)

print(workout_performance)
```

**მიგნებები:**

ანალიზმა დაგვანახა, რომ მომხმარებლები ყველაზე დიდ დროს უთმობენ იოგას, ხოლო ყველაზე მეტი კალორია დაწვეს მორბენლებმა.

**რეკომენდაციები:**

იმისთვის, თუ მომხმარებლეს სურს მეტი კალორიის დაწვა, რეკომენდაციის სახით ვურჩევთ, რომ მისთვის კარგი აქტივობა იქნებოდა სირბილი.

**4. მომხმარებელთა კმაყოფილება**

ახლა ვნახოთ, როგორ აფასებენ მომხმარებლები სხვადასხვა ტიპის ვარჯიშებს:


```python
workout_ratings = df.groupby('ვარჯიშის ტიპი')['მომხმარებლის შეფასება'].mean()

workout_ratings = workout_ratings.sort_values(ascending=False)

print(workout_ratings)
```

**დაკვირვებები:**

მოცემული ინფორმაციის საფუძველზე, ვასკვნით, რომ ყველაზე მაღალი რეიტინგი აქვს ველოსპორტს, ხოლო ყველაზე დაბალი რეიტინგი ძალოსნობას აქვს.

**შეხედულება:**

მოცემული შედეგები მეტყველებს სპორტის სირთულეებზე, რადგან ველომოყვარული ადამიანი ბევრად მარტივად იკმაყოფილებს თავის სურვილებს, ველოსიპედით სეირნობისას, ვიდრე ძალოსანი.

**5. ვარჯიშის ხანგრძლივობისა და კალორიების წვის კორელაცია**

განვიხილოთ, არის თუ არა კავშირი ვარჯიშის ხანგრძლივობასა და დამწვარ კალორიებს შორის:

```python
correlation = df['ხანგრძლივობა'].corr(df['დამწვარი კალორიები'])

print(f"კორელაცია ხანგრძლივობასა და დამწვარ კალორიებს შორის: {correlation:.2f}")
```

**ანალიზი:**

კორელაცია ხანგრძლივობასა და დამწვარ კალორიებს შორის უდრი 0.86. მოცემული შედეგი მეტყველებს იმაზე რომ უფრო ხანგრძლივი ვარჯიში დაკავშირებულია მაღალ კალორიულ ხარჯებთან.

**სტრატეგია:**

ამ კორელაციის გათვალისწინებით აპლიკაციამ ყურადღება უნდა გაამახვილოს მომხმარებელთა წახალისებაზე, რომ გააუმჯობესონ ხანგრძლივობა.

**6. ყოველთვიური ჩართულობის ტენდენციები**

განვიხილოთ, როგორ მერყეობს მომხმარებელთა ჩართულობა წლის განმავლობაში:

```python
df['თარიღი'] = pd.to_datetime(df['თარიღი'])

df['თვე'] = df['თარიღი'].dt.to_period('M')

monthly_engagement = df.groupby('თვე').agg(
    სესია=('ვარჯიშის ტიპი', 'count'),
    საშუალო_ხანგრძლივობა=('ხანგრძლივობა', 'mean'),
    საშუალო_დამწვარი_კალორიები=('დამწვარი კალორიები', 'mean')
)

monthly_engagement.columns = ['სესიები', 'საშუალო ხანგრძლივობა', 'საშუალო კალორიები']
monthly_engagement = monthly_engagement.sort_values(by='სესიები', ascending=False)

print(monthly_engagement)

best_month = monthly_engagement.index[0]
worst_month = monthly_engagement.index[-1]

print(f"\nსაუკეთესო თვე: {best_month}")
print(f"ყველაზე ცუდი თვე: {worst_month}")
```

**დაკვირვებები:**

  ანალიზის შედეგად ვასკვნით, რომ იანვარში ყველაზე მეტი ადამიანი იწყებს ვარჯიშს,
  ხოლო წლის მიწურულს, დეკემბერში, ყველაზე ცოტა სესია ფიქსირდება.

**შეხედულება:**

  ვფიქრობ, მოცემული ანალიზის გამომწვევი მიზანია ახალი წლის საწყისი, რადგან ამ დროს ადამიანები ბევრად მოტივირებულები არიან, ამასთან ერთად მრავალი დღესასწაულის შედეგად მომატებულ კილოგრამების დაკლებას ცდილობენ.

**სტრატეგია:**

წამახალისებელი სტრატეგია წლის საწყის თვეებში შეიძლება იყოს აბონიმენტის ფასდაკლება, საჩუქრები და ა.შ.

**7. მომხმარებელთა თანმიმდევრულობის ანალიზი**

გავაანალიზოთ მომხმარებელთა ვარჯიშის სიხშირეთა განაწილება:

```python
user_frequency = df.groupby('მომხმარებლის ID')['თარიღი'].count()
frequency_stats = user_frequency.describe()
frequency_quantiles = user_frequency.quantile([0.25, 0.5, 0.75])

print("ვარჯიშის სიხშირის სტატისტიკა:")
print(frequency_stats)
print("\nვარჯიშის სიხშირის კვანტილები:")
print(frequency_quantiles)
```

**ანალიზი:**

ტრენინგის საშუალო სიხშირე წელიწადში საშუალოდ 47.75 სესიაზე მიუთითებს, ხოლო სტანდარტული გადახრა უდრის 15.7-ს.

მომხმარებელთა 25% ატარებს 36 სესიან ან ნაკლებს წეილიწადში, ხოლო 75% აქვს 57 ან ნაკლები სესია წელიწადში.

მედიანური სიხშირე არის 45 სესიის ტოლი, რაც გვაჩვენებს, რომ მომხმარებელთა ნახევარი ვარჯიშობს ამ ოდენობაზე ნაკლებს ან ტოლს.


**რეკომენდაცია:**

ნაკლებად აქტიური მომხმარებლების ჩართულობისთვის უნდა მოქმედებდეს წამახალისებელი ქმედებები, მაგალითად, თუ მომხმარებელს არ უვარჯიშია 3 თვე, აპლიკაციის სახით შეიძლება მიუვიდეს შეტყობინება ფასდაკლების შესახებ, წამახალისებელი ასევე შეიძლება იყოს სხვა ნებისმიერი საჩუქარი, რომელსაც მომხმარებელი ვარჯიშის დაწყებისას მიიღებს.
